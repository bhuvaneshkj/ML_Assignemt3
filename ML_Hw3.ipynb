```python
clf = DecisionTreeClassifier(criterion="entropy")

# Train Decision Tree Classifer
clf = clf.fit(x_train_df2,y_train_df2)

#Predict the response for test dataset
y_pred = clf.predict(x_test_df2)
y_pred
```
